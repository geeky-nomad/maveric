```python
import pandas as pd

# Load the dataset
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)
df = pd.read_csv('FashionDataset.csv')

# Display the first few rows of the dataframe
df.head(n=10)
```

```python
# extract category from Details column
import re
def extract_category(details):
    category = re.search(r'womens\s+(.*)', details, re.IGNORECASE)
    if category:
        return category.group(1)
    return None

# Apply the function to the 'Details' column and create/update the 'Category' column
df['Category'] = df['Details'].apply(extract_category)

# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values (if any)
df.dropna(inplace=True)
df.to_csv('update_fashion.csv', index=False)
```
