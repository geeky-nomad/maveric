```python
import re
import nltk
import pandas as pd
# load the csv
df = pd.read_csv('update_fashion.csv')
df.head(2)
```

```python
# Tokenization -> Details
def tokenization(text):
    tokens = text.split()
    return tokens

stopwords = nltk.corpus.stopwords.words('english')

def remove_stopword(tokenize_item):
    item = [text for text in tokenize_item if text not in stopwords]
    return item

# Apply both functions in a single line
df['Tokenized_Details'] = df['Details'].apply(lambda x: remove_stopword(tokenization(x)))

# now finally saving it in new csv
df.to_csv('nlp_fashion.csv', index=False)
```
