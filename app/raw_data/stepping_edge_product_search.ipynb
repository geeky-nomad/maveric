```python
# import necessary libraries
import pandas as pd
import numpy as nm

# set the loading/representation setting
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# loading the data
data_to_load = 1000
df = pd.read_csv('export_catalog_product_20240508_060248.csv', nrows=data_to_load)
# renaming the column for searching through sub-category wise 
df.rename(columns={'attribute_set_code': 'sub-category'}, inplace=True)
df['sub-category'] = df['sub-category'].str.replace('RS', '').str.strip()
# renaming the column categories to category to enabling search through category wise
df.rename(columns={'categories': 'category'}, inplace=True)
df['category'] = df['category'].str.extract(r'Default Category\/([^,]*)')
```
