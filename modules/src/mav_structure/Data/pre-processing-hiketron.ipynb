```python
def find_phrase_indices(sentences, phrase, label):
    results = []
    for sentence in sentences:
        start_index = sentence.find(phrase)
        if start_index != -1:
            end_index = start_index + len(phrase)
            entities = [(start_index, end_index, label)]
            results.append((sentence, {'entities': entities}))
        else:
            results.append((sentence, {'entities': []}))  # No entities found
    return results

# Example usage
sentences = [
    'gift cards',
    'do you also have git cards?',
    'show me gift cards'
]
phrase = "gift cards"
label = "Gift Card"
indices = find_phrase_indices(sentences, phrase, label)
for result in indices:
    print(result)
```
